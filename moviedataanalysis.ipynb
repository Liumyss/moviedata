### **9. Display title of the movie having runtime >= 200 minutes**

```python
data['Runtime'] = data['Runtime'].str.strip(" min")
data['Runtime'] = pd.to_numeric(data['Runtime'])
```

```python
data[data['Runtime']>=200]['Title']
```

### **10. Which year has the highest average voting?

```python
data.groupby('Year')['Votes'].mean().sort_values(ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(80, 60))
sns.barplot(x='Year', y='Votes', data=data, ax=ax)
plt.title("Votes By Year")
plt.show()
```

### **11. Which year has the highest average revenue?**

```python
data['Gross'] = data['Gross'].str.replace(',', '')
data['Gross'] = pd.to_numeric(data['Gross'])

# define a lambda function to round a value to one decimal place
round_to_millions = lambda x: round(x / 100000, 2)

# apply the lambda function to a specific column of the DataFrame
data['Gross'] = data['Gross'].apply(round_to_millions)
```

```python
data.groupby('Year')['Gross'].mean().sort_values(ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(80, 70))
data = data.dropna(subset=['Gross'])
sns.barplot(x='Year', y='Gross', data=data, ax=ax)
plt.title("Revenue By Year")
plt.show()
```
